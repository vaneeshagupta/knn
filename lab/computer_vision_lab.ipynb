````python
#1
from keras.datasets import mnist
df = mnist.load_data('minst.db')
train,test = df
X_train, y_train = train
X_test, y_test = test

import matplotlib.pyplot as plt
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000)
for i in range(5): 
    print(y_test[i],'\n') # Print the label
    print(X_test[i],'\n') # Print the matrix of values
    plt.contourf(np.rot90(X_test[i].transpose())) # Make a contour plot of the matrix values
    plt.show()
````

````python
#2
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print(X_train[0].shape)
print(X_test[0].shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
````

````python
#3
X_train_reshaped = X_train.reshape(X_train.shape[0], 784)
X_test_reshaped = X_test.reshape(X_test.shape[0], 784)
print("New shape of X_train_flat:", X_train_reshaped.shape)
print("New shape of X_test_flat:", X_test_reshaped.shape)

````

````python
#4
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt

k_bar = 20
accuracy = []

for k in range(1, k_bar + 1):  # Loop through k values from 1 to k_bar
    print(f"Current k: {k}")
    knn = KNeighborsClassifier(n_neighbors=k)
    predictor = knn.fit(X_train_reshaped, y_train)  # Fit model
    accuracy.append(predictor.score(X_test_reshaped, y_test))  # Store accuracy

# Find optimal k
accuracy_max = np.max(accuracy)
max_index = np.where(accuracy == accuracy_max)[0][0]  # Ensure single index selection
k_star = max_index + 1  # k values start at 1, not 0
print(f"Optimal k: {k_star}")

# Plot accuracy vs. k
plt.plot(np.arange(1, k_bar + 1), accuracy)
plt.xlabel("Number of Neighbors (k)")
plt.ylabel("Accuracy")
plt.title(f"Optimal k: {k_star}")
plt.grid()
plt.show()

````

The optimal k is 3.

````python
#5
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd

knn = KNeighborsClassifier(n_neighbors=3)
predictor = knn.fit(X_train_reshaped, y_train)  
y_hat = predictor.predict(X_test_reshaped)
accuracy = knn.score(X_test_reshaped, y_test)  
print('Accuracy:', accuracy)
confusion_matrix_df = pd.crosstab(y_test, y_hat, rownames=["Actual"], colnames=["Predicted"])
print(confusion_matrix_df)

````

The kNN classifier with k=3 achieves an accuracy of 97.05% on the test set, meaning it correctly identifies digits in 97.05% of cases. This indicates that the model is highly effective at recognizing handwritten digits, but there are still some misclassifications. The confusion matrix reveals that certain digits are frequently mistaken for others, highlighting areas where the model struggles (ex. 4 for 9)

The most common mistakes occur between similar-looking digits. For example

- 7 is often confused with 1 (21 times)  This is likely because some handwritten 7s look similar to 1s, especially if they lack a distinguishing horizontal stroke.
- 4 is confused with 9 (19 times)  Handwritten 4s sometimes resemble 9s, particularly if the top part is not clearly written.
- 2 is misclassified as 7 (13 times)  This could be due to the curved structure of both numbers.
- 3 is confused with 5 (13 times)  Some writing styles make 3 and 5 look similar, particularly when the bottom of the 3 is slightly curved inward.
- 8 is confused with 3 and 5  The number 8 is sometimes misclassified as 3 (16 times) or 5 (11 times), possibly due to variations in how loops are written.

These errors suggest that digits with similar structures or handwriting styles can be misclassified. The most common confusions tend to occur between numbers that share curves or strokes which I find interesting.

To represent a color photo as a matrix, we break the image into pixels, each having an RGB (Red, Green, Blue) value. This can be represented as three separate 2D matrices, one for each color channel (Red, Green, Blue). Each matrix would have dimensions 2828 (if the image is 28x28 pixels), where each value represents the intensity of that particular color at each pixel. Next, we reshape each of these 28x28 matrices into a 1D vector, and then concatenate them side by side into a single long row, creating tabular data. This flattened vector can then be treated as a row in a dataset, where each column represents the intensity of a specific color component for a particular pixel. This allows machine learning algorithms to process the image data just like any other tabular data. So in short: you create three separate 28x28 matrices for Red, Green, and Blue intensities, flatten them, and combine them into one long vector.
